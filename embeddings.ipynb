```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent=[  'the glass of milk',
      'the glass of juice',
      'the cup of tea',
      'I am a good boy',
      'I am a good developer',
      'understand the meaning of words',
      'your videos are good']
```

```python
voc_size=10000
sent
```

```python
one_hot_rep=[one_hot(word,voc_size) for word in sent]
one_hot_rep
```

```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
import pandas as pd
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
dim=10
```

```python
# Create model with explicit input shape
model = Sequential()
model.add(Embedding(input_dim=voc_size, output_dim=dim, input_length=sent_length))

# Force model to build by providing sample input shape
model.build(input_shape=(None, sent_length))

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
